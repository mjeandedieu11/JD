```python
# Generator function to generate Fibonacci numbers
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Create Fibonacci generator
fib_gen = fibonacci()

# Filter even Fibonacci numbers using lambda and filter()
even_fib = filter(lambda x: x % 2 == 0, fib_gen)

# Print the first 10 even Fibonacci numbers
count = 0
for num in even_fib:
    print(num)
    count += 1
    if count == 10:
        break
```

```python
def count_up_to(max_value):
    current = 1
    while current <= max_value:
        yield current
        current += 1
# Using the generator
counter = count_up_to(5)
for number in counter:
    print(number)
```
